```python
# Erster Datensatz
import pandas as pd
import requests
url = "https://www.bibsonomy.org/json/search/apollo?items=1000"
result = requests.get(url) 
data = result.json()
```

```python
data.keys()
```

```python
df_bibsonomy = pd.DataFrame(data["items"])
```

```python
df_bibsonomy
```

```python
## zweiter Datensatz
url = "https://www.bibsonomy.org/json/search/spacecraft?items=1000"
result = requests.get(url) 
data1 = result.json()
```

```python
df_bibsonomy1 = pd.DataFrame(data1["items"])
```

```python
df_bibsonomy1
```

```python
## Dritter Datzensatz

url = "https://www.bibsonomy.org/json/search/black%20holes?items=1000"
result = requests.get(url) 
data2 = result.json()
```

```python
df_bibsonomy2 = pd.DataFrame(data2["items"])
```

```python
df_bibsonomy2
```

```python
# Datensätze kombinieren
df_all = pd.concat([df_bibsonomy, df_bibsonomy1, df_bibsonomy2])
```

```python
df_all
```

```python
# Nach nur type = Publication filtern

df_all.loc[df_all.type == "Publication"]
```

```python
# Ansicht nur ausgewählter Spalten
df_all[["label", "author", "year", "type"]]
```

```python
# Ansicht nur der Spalte type
df_all["type"]
```

```python
# Gefilterte Ansicht als Datensatz speichern (gleicher Name = überschreibt die alte Datei)
df_all = df_all.loc[df_all.type == "Publication"]
```

```python
df_all["type"]
```

```python
df_all[["label", "author", "year", "type"]]
```

```python
# Welche colmuns gibt es
df_all.columns
```

```python
# Wieder ausgewählte Ansicht von Spalten
df_all[["label", "author", "affiliation", "year", "type", "id"]]
```

```python
# Speichern der gefilterten Ansicht
df_all = df_all[["label", "author", "affiliation", "year", "type", "id"]]
```

```python
df_all[["affiliation"]]
```

```python
df_all
```

```python
# Wo steht in affiliation überhaupt was drin:

df_all.loc[df_all["affiliation"].isnull() == False]
```

```python
df_all[["id"]]
```

```python
df_all.id
```

```python
# alles rausfiltern, was bei author keinen Eintrag hat (= alles behalten, was einen Eintrag hat)

df_all = df_all.loc[df_all["author"].isnull() == False]
```

```python
df_all
```

```python
# gibt es Einträge ohne Titel (label) -> Nein

df_all.loc[df_all["label"].isnull() == True]
```

```python
# gibt es Einträge ohne Jahr (year) -> Nein

df_all.loc[df_all["year"].isnull() == True]
```

```python
# Dublettenprüfung über den Titel

df_all = df_all.drop_duplicates(subset = "label", keep = "first")
```

```python
df_all
```

```python
# Jahreseintrag bereinigen
df_all.year
```

```python
# Gibt es unsaubere Einträge? -> ja
df_all["year"].unique()
```

```python
# Welchen Datentyp hat das Jahr? -> string
df_all.loc[df_all["year"] > 2010]
```

```python
# Aussortieren aller unsauberen Zahlen, bzw. Abfrage, in welcher Zeile eine unsaubere Zahl steht in neuer Spalte "year_correct"
df_all["year_correct"] = df_all.apply(lambda row: row.year.isdigit(), axis=1)
```

```python
# Anzeige aller Einträge mit unsauberer Zahl, die nicht in int umgewandelt werden können und speichern der Gegenprobe
df_all.loc[df_all.year_correct == False]
```

```python
df_all = df_all.loc[df_all.year_correct == True]
```

```python
df_all
```

```python
# Umwandeln der Zahlen in int und speichern in neuer Spalte publication_year
df_all["publication_year"] = df_all["year"].astype(int)
```

```python
df_all
```

```python
# Löschen überflüssiger Spalten
df_all = df_all.drop(columns=['year', 'year_correct'])
```

```python
df_all
```

```python
# Spalten umbenennen
df_all = df_all.rename(columns={"label": "title", "author": "authors", "id": "source"})
```
