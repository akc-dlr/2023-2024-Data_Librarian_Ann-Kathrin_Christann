```python
## Export von Datens√§tzen aus Bibsonomy
# Export 1, Stichwort "space travel" - dataset_space
```

```python
import pandas as pd
```

```python
import requests
url = "https://www.bibsonomy.org/json/search/space%20travel?items=1000"
result = requests.get(url) 
dataset_space = result.json() 
```

```python
dataset_space
```

```python
# Export 2, Stichwort "moon"
```

```python
url = "https://www.bibsonomy.org/json/search/moon?items=1000"
result = requests.get(url) 
dataset_moon = result.json() 
```

```python
# Export 3, Stichwort "lasers"
```

```python
url = "https://www.bibsonomy.org/json/search/lasers?items=1000"
result = requests.get(url) 
dataset_lasers = result.json() 
```

```python
## Die drei Datensets zusammenf√ºhren
```

```python
df_spacemoonlaser = pd.merge(pd.merge(dataset_space, dataset_moon, dataset_lasers))
```
